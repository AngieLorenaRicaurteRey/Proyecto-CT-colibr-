```python
import os
import numpy as np
import tifffile as tiff
from skimage import exposure

ruta_origen = r"c:\Users\a.ricaurte\Desktop\nuevas_proyecciones_colibri"
ruta_salida = r"c:\Users\a.ricaurte\Desktop\contraste_muy_alto_pa_reconstruir"
os.makedirs(ruta_salida, exist_ok=True)

for nombre_archivo in os.listdir(ruta_origen):
    if nombre_archivo.lower().endswith(('.tif', '.tiff')):
        ruta_completa = os.path.join(ruta_origen, nombre_archivo)
        try:

            imagen = tiff.imread(ruta_completa).astype(np.float32)

   
            p2, p98 = np.percentile(imagen, (2, 98))
            imagen_ventaneada = np.clip((imagen - p2) / (p98 - p2), 0, 1)

        
            imagen_ecualizada = exposure.equalize_adapthist(imagen_ventaneada, clip_limit=0.06)

        
            imagen_uint16 = (imagen_ecualizada * 65535).astype(np.uint16)

        
            ruta_guardar = os.path.join(ruta_salida, nombre_archivo)
            tiff.imwrite(ruta_guardar, imagen_uint16)

            print(f"Procesada y guardada (uint16): {nombre_archivo}")

        except Exception as e:
            print(f"Error al procesar {nombre_archivo}: {e}")

print("Proceso completado.")

```
