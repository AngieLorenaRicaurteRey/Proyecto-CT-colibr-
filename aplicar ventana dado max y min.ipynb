```python
import os
import numpy as np
from tifffile import imread, imwrite
intensity_min =  17550
intensity_max = 35485
ruta_entrada = r"C:\Users\a.ricaurte\Desktop\tajadas_seleccionadas_dos"
ruta_salida = r"C:\Users\a.ricaurte\Desktop\v2"
os.makedirs(ruta_salida, exist_ok=True)

def apply_windowing(image, intensity_min, intensity_max):

    windowed_img = np.clip(image, intensity_min, intensity_max) 
    windowed_img = ((windowed_img - intensity_min) / (intensity_max - intensity_min)) * 65535 
    return windowed_img.astype(np.uint16)


for filename in os.listdir(ruta_entrada):
    if filename.endswith(".tif") or filename.endswith(".tiff"):
        img_path = os.path.join(ruta_entrada, filename)
        output_path = os.path.join(ruta_salida, filename)

        img = imread(img_path) 
        windowed_img = apply_windowing(img, intensity_min, intensity_max)  
        imwrite(output_path, windowed_img)  

        print(f"Procesado: {filename}")

print("Procesamiento completado.")



#(2,98) 
# Mínima intensidad gris: 6308 en (910, 618)
# Máxima intensidad gris: 35485 en (673, 945)



```
