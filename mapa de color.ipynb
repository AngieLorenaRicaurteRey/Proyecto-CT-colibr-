```python
import matplotlib.pyplot as plt
from skimage.io import imread
from skimage import exposure
import numpy as np


img = imread(r"c:\Users\a.ricaurte\Desktop\colibri_0900.tif")


print("Rango de intensidades:", img.min(), "a", img.max())


p2, p98 = np.percentile(img, (2, 98))
img_rescaled = exposure.rescale_intensity(img, in_range=(p2, p98), out_range=(0, 255)).astype(np.uint8)


plt.figure(figsize=(10, 8))
plt.imshow(img_rescaled, cmap='jet') 
plt.colorbar(label="Intensidad (rescalada)")
plt.title("Colibri con colormap 'jet'")
plt.axis('off')
plt.show()
print(p2,p98)
```
